```python
import numpy as np
```

```python
# n-identity matrix
np.identity(n=8, dtype='int64')
```

```python
 np.identity(n=10, dtype='float32') # dtype = data type, can set size
```

```python
# eye matrix
np.eye(8, k=2, dtype=int) # k: position of column, count from row 0
```

```python
# linspace
x = np.linspace(0, 1, 11)  # note the {stop} section is inclusive
x
```

```python
# if you want to create an 'empty' array stick to np.zeros - if you forget to initialize a number it could be a rounding error
y = np.empty(3)
y
```

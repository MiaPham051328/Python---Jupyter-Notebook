```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

from skimage import data
from skimage.color import rgb2gray

matplotlib.rcParams['font.size'] = 18

images = ('astronaut',
          'binary_blobs',
          'brick',
          'colorwheel',
          'camera',
          'cat',
          'checkerboard',
          'clock',
          'coffee',
          'coins',
          'eagle',
          'grass',
          'gravel',
          'horse',
          'logo',
          'page',
          'text',
          'rocket',
          )
img = data.astronaut()
plt.figure()
plt.imshow(img); # ~img: change color
```

```python
img.shape
# img.ndim
```

```python
img = rgb2gray(img)
plt.imshow(img,cmap=plt.cm.gray);
```
